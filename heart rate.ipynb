```python
def printhr():
 
    try:
        result = theCursor.execute("SELECT time, hr FROM heartrate")
        with open("out20.csv","w") as f:
            wr = csv.writer(f,delimiter='\n')
            wr.writerow(result)
            print("Check Your Directory for the CSV")  
            f.close()
            
    except sqlite3.OperationalError:
        print("The Table Doesn't Exist")
 
    except:
        print("Couldn't Retrieve Data From Database")
 
db_conn = sqlite3.connect('db.sqlite')
print("Database Created")
theCursor = db_conn.cursor()
printhr()
```

```python
import pandas as pd
import numpy as np
x=pd.read_csv('out20.csv')
```

```python
x.columns=["time stamp","heart beat"]
```

```python
x['time stamp'] = x['time stamp'].apply(lambda x: np.int(x.replace("(","")))
x['heart beat'] = x['heart beat'].apply(lambda x: np.int(x.replace(")","")))
```

```python
x['time stamp'] = pd.to_datetime(x['time stamp'],unit='s')
```
